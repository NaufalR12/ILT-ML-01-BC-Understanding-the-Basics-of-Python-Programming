### Procedure
In KBBI, the word procedure has the meaning of a stage of activity to complete an activity. This is the same as procedure as a subprogram which is a grouping of instructions that are often used in programs.

Unlike functions, procedures do not require input or output parameters and can be viewed as functions that do not produce values. In Python, a procedure is defined by an expressionless return or a value generated at the end of the function.

Conceptually, the image above is the basic framework of a procedure in Python. At first glance, it is very similar to a function, except that we don't define returns and even return values.

### Method

```python
def my_decorator(func):
    def wrapper():
        print("Sebelum fungsi dieksekusi.")
        func()
        print("Setelah fungsi dieksekusi.")
    return wrapper

# Dekorasi fungsi dengan decorator
@my_decorator
def say_hello():
    print("Hello, world!")
```

```python
class Person:
    population = 0

    def __init__(self, name, age):
        self.name = name
        self.age = age
        Person.population += 1

    @classmethod
    def from_birth_year(cls, name, birth_year):
        current_year = 2024  # Misalkan tahun sekarang adalah 2024
        age = current_year - birth_year
        return cls(name, age)

    @classmethod
    def get_population(cls):
        return cls.population

# Membuat instance menggunakan metode alternatif
person1 = Person("Alice", 30)
person2 = Person.from_birth_year("Bob", 1994)

print(person1.name, person1.age)  # Output: Alice 30
print(person2.name, person2.age)  # Output: Bob 30

# Mengakses variabel kelas
print(Person.get_population())  # Output: 2
```

### Inheritance

```python
class Mobil:
    def __init__(self, warna, merek, kecepatan):
        self.warna = warna
        self.merek = merek
        self.kecepatan = kecepatan

    def tambah_kecepatan(self):
        self.kecepatan += 10


class MobilSport(Mobil):
    def turbo(self):
        self.kecepatan += 50

# Kelas Mobil Dasar
mobil_1 = Mobil("Merah", "DicodingCar", 160)
print(mobil_1.kecepatan)

# Kelas Mobil Sport
mobil_sport_1 = MobilSport("Hitam", "DicodingSportCar", 160)
print(mobil_sport_1.kecepatan)
mobil_sport_1.turbo()
print(mobil_sport_1.kecepatan)
```

### Override

```python
class Mobil:
    def __init__(self, warna, merek, kecepatan):
        self.warna = warna
        self.merek = merek
        self.kecepatan = kecepatan

    def tambah_kecepatan(self):     # tambah_kecepatan
        self.kecepatan += 10

class MobilSport(Mobil):
    def turbo(self):
        self.kecepatan += 50

    def tambah_kecepatan(self):     # tambah_kecepatan
        self.kecepatan += 30

# Kelas Mobil Sport
mobil_sport_1 = MobilSport("Hitam", "DicodingSportCar", 160)
print(mobil_sport_1.kecepatan)
mobil_sport_1.tambah_kecepatan()     # Memanggil metode baru tambah kecepatan()
print(mobil_sport_1.kecepatan)
```

### Super

```python
class Mobil:
    def __init__(self, warna, merek, kecepatan):
        self.warna = warna
        self.merek = merek
        self.kecepatan = kecepatan

    def tambah_kecepatan(self):
        self.kecepatan += 10


class MobilSport(Mobil):
    def turbo(self):
        self.kecepatan += 50

    def tambah_kecepatan(self):
        super().tambah_kecepatan()
        print("Kecepatan Anda meningkat! Hati-Hati!")

# Kelas Mobil Sport
mobil_sport_1 = MobilSport("Hitam", "DicodingSportCar", 160)
mobil_sport_1.tambah_kecepatan()     # Memanggil metode baru tambah kecepatan()
print(mobil_sport_1.kecepatan)
```

```python
# Definisikan fungsi yang akan diuji
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        raise ValueError("Cannot divide by zero")
    return a / b
```

```python
import unittest

class TestMathOperations(unittest.TestCase):

    def test_add(self):
        self.assertEqual(add(10, 5), 15)
        self.assertEqual(add(-1, 1), 0)
        self.assertEqual(add(-1, -1), -2)
        self.assertEqual(add(-1, -1), -3)

    def test_subtract(self):
        self.assertEqual(subtract(10, 5), 5)
        self.assertEqual(subtract(-1, 1), -2)
        self.assertEqual(subtract(-1, -1), 0)

    def test_multiply(self):
        self.assertEqual(multiply(10, 5), 50)
        self.assertEqual(multiply(-1, 1), -1)
        self.assertEqual(multiply(-1, -1), 1)

    def test_divide(self):
        self.assertEqual(divide(10, 5), 2)
        self.assertEqual(divide(-1, 1), -1)
        self.assertEqual(divide(-1, -1), 1)
        with self.assertRaises(ValueError):
            divide(10, 0)

if __name__ == '__main__':
    # Create a new TestSuite to isolate tests in this cell
    suite = unittest.TestSuite()
    suite.addTest(unittest.makeSuite(TestMathOperations))
    unittest.TextTestRunner(verbosity=2).run(suite)
```

```python
class TestStringMethods(unittest.TestCase):

    def test_upper(self):
        self.assertEqual('foo'.upper(), 'FOO')

    def test_isupper(self):
        self.assertTrue('FOO'.isupper())
        self.assertFalse('Foo'.isupper())

    def test_split(self):
        s = 'hello world'
        self.assertEqual(s.split(), ['hello', 'world'])
        # check that s.split fails when the separator is not a string
        with self.assertRaises(TypeError):
            s.split(2)

if __name__ == '__main__':
    # Create a new TestSuite to isolate tests in this cell
    suite = unittest.TestSuite()
    suite.addTest(unittest.makeSuite(TestStringMethods))
    unittest.TextTestRunner(verbosity=2).run(suite)
```

```python
import unittest

def koneksi_ke_db():
    print("[terhubung ke db]")

def putus_koneksi_db(db):
    print("[tidak terhubung ke db {}]".format(db))

class User:
    username = ""
    aktif = False

    def __init__(self, db, username):  # using db sample
        self.username = username

    def set_aktif(self):
        self.aktif = True

class TestUser(unittest.TestCase):
    # Test Case 1
    def test_user_default_not_active(self):
        db = koneksi_ke_db()
        dicoding = User(db, "dicoding")
        self.assertFalse(dicoding.aktif)  # tidak aktif secara default
        putus_koneksi_db(db)

    # Test Case 2
    def test_user_is_active(self):
        db = koneksi_ke_db()
        dicoding = User(db, "dicoding")
        dicoding.set_aktif()  # aktifkan user baru
        self.assertTrue(dicoding.aktif)
        putus_koneksi_db(db)

if __name__ == "__main__":
    # Create a new TestSuite to isolate tests in this cell
    suite = unittest.TestSuite()
    suite.addTest(unittest.makeSuite(TestUser))
    unittest.TextTestRunner(verbosity=2).run(suite)
```

```python
import unittest
```

```python
class Portfolio:
    def __init__(self):
        self.stocks = []

    def buy(self, stock_symbol, quantity, price):
        self.stocks.append((stock_symbol, quantity, price))

    def cost(self):
        return sum(quantity * price for _, quantity, price in self.stocks)

class PortfolioTest(unittest.TestCase):
    def test_buy_one_stock(self):
        p = Portfolio()
        p.buy("IBM", 100, 176.48)
        self.assertEqual(p.cost(), 17648.00)

    def test_buy_multiple_stocks(self):
        p = Portfolio()
        p.buy("AAPL", 50, 120.00)
        p.buy("GOOGL", 20, 1500.00)
        self.assertEqual(p.cost(), 36000.00)

# Create a test suite to isolate the tests
def suite():
    suite = unittest.TestSuite()
    suite.addTest(PortfolioTest('test_buy_one_stock'))
    suite.addTest(PortfolioTest('test_buy_multiple_stocks'))
    return suite

# Run the test suite
if __name__ == "__main__":
    runner = unittest.TextTestRunner(verbosity=2)
    runner.run(suite())
```

```python
import sqlite3

# Create a function to insert stock purchase into the database
def insert_stock_purchase(conn, stock_symbol, quantity, price):
    total_cost = quantity * price
    with conn:
        conn.execute(
            "INSERT INTO stock_purchases (stock_symbol, quantity, price, total_cost) VALUES (?, ?, ?, ?)",
            (stock_symbol, quantity, price, total_cost)
        )

# Function to set up an in-memory SQLite database for testing
def create_test_db():
    conn = sqlite3.connect(':memory:')  # In-memory database for testing
    conn.execute('''
        CREATE TABLE stock_purchases (
            id INTEGER PRIMARY KEY,
            stock_symbol TEXT NOT NULL,
            quantity INTEGER NOT NULL,
            price REAL NOT NULL,
            total_cost REAL NOT NULL
        )
    ''')
    return conn
```

```python
class TestSQL(unittest.TestCase):

    def setUp(self):
        # Set up an in-memory SQLite database
        self.conn = create_test_db()

    def tearDown(self):
        # Close the connection after each test
        self.conn.close()

    def test_insert_stock_purchase(self):
        # Test inserting one stock purchase into the database
        insert_stock_purchase(self.conn, "AAPL", 10, 150.0)
        cursor = self.conn.cursor()
        cursor.execute("SELECT * FROM stock_purchases")
        rows = cursor.fetchall()

        # Check if one record was inserted
        self.assertEqual(len(rows), 1)
        # Verify the content of the inserted row
        self.assertEqual(rows[0][1], "AAPL")
        self.assertEqual(rows[0][2], 10)
        self.assertEqual(rows[0][3], 150.0)
        self.assertEqual(rows[0][4], 1500.0)

    def test_multiple_stock_purchases(self):
        # Test inserting multiple stock purchases
        insert_stock_purchase(self.conn, "AAPL", 10, 150.0)
        insert_stock_purchase(self.conn, "IBM", 5, 100.0)
        cursor = self.conn.cursor()
        cursor.execute("SELECT * FROM stock_purchases")
        rows = cursor.fetchall()

        # Check if two records were inserted
        self.assertEqual(len(rows), 2)

# Create a test suite to isolate the SQL tests
def sql_suite():
    suite = unittest.TestSuite()
    suite.addTest(TestSQL('test_insert_stock_purchase'))
    suite.addTest(TestSQL('test_multiple_stock_purchases'))
    return suite

# Run the test suite
if __name__ == "__main__":
    runner = unittest.TextTestRunner(verbosity=2)
    runner.run(sql_suite())
```
